```python
import boto3
from datasets import load_dataset
import json
import os

# Load the dataset from Hugging Face
dataset = load_dataset("ChristophSchuhmann/improved_aesthetics_6plus")

# Create an S3 client
s3 = boto3.client('s3',
                  aws_access_key_id = os.environ.get('AWS_ACCESS_KEY'),
                  aws_secret_access_key = os.environ.get('AWS_SECRET_KEY')
                  )

# Set the S3 bucket name and key
bucket_name = 'laion-6plus'
key = 'improved_aesthetics_6plus'

# Upload each file to S3
# Next time, load in a parallelized manner
for split_name, split in dataset.items():
    for i, example in enumerate(split):
        filename = f'{split_name}/{i}.json'
        s3.put_object(
            Bucket=bucket_name,
            Key=f'{key}/{filename}',
            Body=json.dumps(example, indent=4).encode('utf-8')
        )

print('Upload complete!')
```
